```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import smtplib
from email.mime.text import MIMEText

#GOOGLE SHEET CONFIG
scope = [
    "https://spreadsheets.google.com/feeds",
    "https://www.googleapis.com/auth/drive"
]

creds = ServiceAccountCredentials.from_json_keyfile_name(
    "credentials.json", scope
)
client = gspread.authorize(creds)

sheet = client.open("DESIGNER_TARGET").sheet1
data = sheet.get_all_records()

#EMAIL CONFIG
FROM_EMAIL = "ravindukavishkaaluvihare@gmail.com"
EMAIL_PASSWORD = "#######"  # Gmail App Password

#RULE-BASED AI MESSAGE
def generate_message(name, target, completed):
    pending = target - completed

    if pending <= 0:
        return f"""
Hi {name},

ðŸŽ‰ Excellent work!

You have successfully completed your weekly target.
Target: {target}
Completed: {completed}

Keep up the great performance!
"""
    elif pending <= 10:
        return f"""
Hi {name},

ðŸ‘ You're almost there!

Target: {target}
Completed: {completed}
Pending: {pending}

Just a little more to go â€” keep pushing!
"""
    else:
        return f"""
Hi {name},

ðŸ“Œ Weekly Target Update

Target: {target}
Completed: {completed}
Pending: {pending}

Stay focused and keep moving forward. You can do it!
"""

#EMAIL FUNCTION
def send_email(to_email, subject, message):
    msg = MIMEText(message)
    msg["Subject"] = subject
    msg["From"] = FROM_EMAIL
    msg["To"] = to_email

    server = smtplib.SMTP_SSL("smtp.gmail.com", 465)
    server.login(FROM_EMAIL, EMAIL_PASSWORD)
    server.send_message(msg)
    server.quit()

    print(f"âœ… Email sent to {to_email}")

#MAIN LOOP 
for row in data:
    name = row["Designer Name"]
    email = row["Email ID"]
    target = int(row["Weekly Target"])
    completed = int(row["Completed"])

    message = generate_message(name, target, completed)
    send_email(email, "Weekly Target Update", message)

print("ðŸŽ¯ All emails sent successfully!")
```
